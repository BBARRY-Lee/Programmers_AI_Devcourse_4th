## 정보를 달라고 요청하기, GET

```python
# requests 라이브러리를 불러온 후, NAVER의 홈 페이지를 요청한 후 응답 받아보기
import requests
res = requests.get("http://www.naver.com")
res # HTTP 응답이 답겨있음 -> Response [200]
```

```python
# Header를 확인해봅시다 : .headers
res.headers
```

```python
# Body를 텍스트 형태로 확인해봅시다 : .text
res.text[:1000] # 1000번째 문자까지만 확인

# 일부분만 원할 경우는 ?
```

## 정보 갱신하는 것을 요청하기, POST

그렇다면 이렇게 항상 정보를 가져오는 경우만 있습니다.

아닙니다. 우리는 우리의 정보를 제공해주면서 서버로하여금 무언가 요청을 진행할 때도 있습니다.
우리가 매일 진행하는, 로그인이 바로 그 예시죠.

이러한 요청을 처리하는 HTTP Method가 있는데, 바로 POST입니다.

POST를 활용하기 위해 다음 사이트를 이용해보도록 하겠습니다: https://webhook.site

```python
# payload와 함께 POST를 보내봅시다 : requests.post()
payload = {"name": "Hello", "age": 13} #json은 파이썬의 딕셔너리와 유사

res = requests.post("https://webhook.site/56702cd0-6b89-47d9-882a-f80f100b92c6", payload)
```

응답의 상태 코드를 통해 응답이 잘 왔는지 확인해볼까요?

```python
# 상태 코드(status code)를 확인해봅시다 : .status_code

res.status_code
# 200이 나오면 잘 된 것
```
