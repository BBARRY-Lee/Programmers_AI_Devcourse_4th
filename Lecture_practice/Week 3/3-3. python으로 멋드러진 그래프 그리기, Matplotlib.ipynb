```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 2. Case Study with Arguments

```python
plt.plot([1, 2, 3, 4, 5]) # 실제 plotting을 하는 함수 -> y = x + 1
# -> y = 1, 2, 3, 4, 5, x = idx
plt.show() # plt 확인
```

### Figsize : Fiqure(도면)의 크기 조절

```python
plt.figure(figsize = (3, 3)) # plotting할 도면을 선언
plt.plot([0, 1, 2, 3, 4])
plt.show()
```

### 2차 함수 그래프 with plot()

```python
# 리스트를 이용해 1차함수 y=x 그려보면 : 
plt.plot([0,1,2,3,4,5])
plt.show()
```

```python
# numpy.array를 이용해 함수 그리기
# y = x^2
x = np.array([1, 2, 3, 4, 5]) # 정의역
y = np.array([1, 4, 9, 16, 25]) # f(x) value
plt.plot(x, y)
plt.show()
```

```python
# np.arange(a, b, c) c = 0.01

x = np.arange(-10, 10, 0.01) # -10~10 까지 0.01씩 증가
plt.plot(x, x**2)
plt.show()
```

```python
# x, y축에 설명 추가하기

x = np.arange(-10, 10, 0.01) 

###
plt.xlabel('x value')
plt.ylabel('f(x) value')
###

plt.plot(x, x**2)
plt.show()
```

```python
# x, y축의 범위 설정하기

x = np.arange(-10, 10, 0.01) 
plt.xlabel('x value')
plt.ylabel('f(x) value')

###
plt.axis([-5, 5,0, 25]) # [x_min, x_max, y_min, y_max]
###

plt.plot(x, x**2)
plt.show()
```

```python
# x, y 축에 눈금 설정하기

x = np.arange(-10, 10, 0.01) 
plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.axis([-5, 5,0, 25]) 

###
plt.xticks([i for i in range(-5, 5, 1)]) # x축 눈금설정 : -5, -4, 3 ...
plt.yticks([i for i in range(0, 27, 3)]) # y축 눈금설정 : 0, 3, 6 ,,,
###

plt.plot(x, x**2)
plt.show()
```

```python
# 그래프 title 설정하기

x = np.arange(-10, 10, 0.01) 
plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.axis([-5, 5,0, 25]) 
plt.xticks([i for i in range(-5, 5, 1)]) 
plt.yticks([i for i in range(0, 27, 3)]) 

###
plt.title('y=x^2 graph')
###

plt.plot(x, x**2)
plt.show()
```

```python
# 범례 설정

x = np.arange(-10, 10, 0.01) 
plt.xlabel('x value')
plt.ylabel('f(x) value')
plt.axis([-5, 5,0, 25]) 
plt.xticks([i for i in range(-5, 5, 1)]) 
plt.yticks([i for i in range(0, 27, 3)]) 
plt.title('y=x^2 graph')

###
plt.plot(x, x**2, label = 'trend') # 순서유의
plt.legend() # 순서유의
### 
plt.show()
```

## III. Matplotlib Case Study

### 꺾은선 그래프 (Plot) -> 시계열 데이터에서 많이 사용

```python
x = np.arange(20) # 0 ~ 19
y = np.random.randint(0, 20, 20) # 0 ~ 20까지 난수를 20번 생성

plt.plot(x, y)

## Extra : y축을 20까지 보이게하고 싶다면? y축을 5단위로 보이게하고 싶다면?
plt.axis([0, 20, 0, 20])
plt.yticks([i for i in range(0, 21, 5)])
plt.show()
```

### 산점도 (Scatter Plot) -> 상관관계 파악

```python
plt.scatter(x, y)
plt.show()
```

### Box plot
- 수치형 데이터에 대한 정보 (Q1, Q2, Q3, min, max)

```python
plt.boxplot((x, y))


# Extra : plot의 title
plt.title('Box plot of x, y')
plt.show()
```

### 막대그래프 (Bar plot)
- 범주형 데이터의 값과 그 크기를 직사각형으로 나타낸 그림

```python
plt.bar(x, y)


# Extra : xtics
plt.xticks(np.arange(0, 20, 1))
plt.show()
```

### cf) histogram
- 도수분포를 직사각형의 막대형태로 나타냄
- 계급으로 나타낸 것이 특징 -> 0, 1, 2가 아닌 0~2까지 범주형 데이터로 구성 후 그림을 그림

```python
plt.hist(y, bins = np.arange(0, 20, 2))


# Extra : xtics를 고쳐보자
plt.xticks(np.arange(0, 20, 2))
plt.show()
```

### 원형 그래프 (Pie CHart)

- 데이터에서 전체에 대한 부분 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해 비율 확인에 용이

```python
z = [100, 300, 200, 400]
plt.pie(z, labels = ['one', 'two', 'three', 'four'])
plt.show()
```

## IV. The 멋진 그래프, Seaborn Case Study

### matplotlib를 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리
- 커널밀도그림
- 카운트그림
- 캣그림
- 스트립그림
- 히트맵

```python
import seaborn as sns
```

### 커널밀도그림 (Kernel Density Plot)
- 히스토그램과 같은 연속적 분포를 곡선화하여 그린 그림
- sns.kdeplot()

```python
# in Hist

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins = x)
plt.show()
```

```python
# kdeplot
#shade = True하면 색을 칠해줌, defalt는 False -> 버전에 따라 shade, fill 다른듯..! 내 버전에서는 fill
sns.kdeplot(y, fill = True)
plt.show()
```

```python
vote_df = pd.DataFrame({'name' : ['Andy', 'Bob', 'Cat'], 'vote' : [True, True, False]})
vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby('vote').count()
vote_count
```

```python
plt.bar(x = [False, True], height = vote_count['name'])
plt.show()
```

```python
# sns의 countplot -> 이 그래프다 더 깔끔하다.

sns.countplot(x = vote_df['vote'])
plt.show()
```

### 캣그림 (Cat Plot)
- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수

```python
covid = pd.read_csv("./country_wise_latest.csv")
covid.head(5)
```

```python
s = sns.catplot(x = 'WHO Region',y = 'Confirmed', data=covid, kind = 'violin') # kind로 여러 형태로 표현 가능, defalt -> kind = strip
s.fig.set_size_inches(10, 6) # x축 글씨가 겹침 -> 사이즈 지정
plt.show()
```

### 스트립 그림 (Strip Plot)
- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프

```python
sns.stripplot(x = 'WHO Region', y = 'Recovered', data = covid)

plt.show()
```

```python
# cf) swarmplot
sns.swarmplot(x = 'WHO Region', y = 'Recovered', data = covid)

plt.show()
# 오류 무시
```

### 히트맵 (Heatmap)
- 데이터의 행렬을 색상으로 표현해주는 그래프

```python
covid.corr()
```

```python
sns.heatmap(covid.corr())
plt.show()
```
