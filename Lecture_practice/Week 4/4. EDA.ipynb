## 3. 데이터의 개별 속성 파악하기

```python
# 생존자, 사망자의 수 확인

titanic_df['Survived'].sum()
# or
titanic_df['Survived'].value_counts()
```

```python
# 생존자, 사망자의 수 Barplot으로 그리기 - sns.countplot()

sns.countplot(x = 'Survived', data = titanic_df)
plt.show()
```

# 2. Pclass

```python
# Pclass에 따른 인원파악 - count()

total = titanic_df[['Pclass', 'Survived']].groupby(['Pclass']).count()
total

# Pclass에 대한 인원 파악 가능
```

```python
# 그렇다면 생존자 인원은? - sum()
surv = titanic_df[['Pclass', 'Survived']].groupby(['Pclass']).sum()
surv
```

```python
# 전체 인원 대비 생존자의 비율
pct = (surv/total)*100
pct # -> 내생각

# 강의
pct2 = titanic_df[['Pclass', 'Survived']].groupby(['Pclass']).mean()
pct2
```

```python
# 히트맵 적용 -> 상관관계 확인 : 밝으면(1) 생존확률이 높고, 어두워질수록(2~3) 낮아짐
sns.heatmap(titanic_df[['Pclass', 'Survived']].groupby(['Pclass']).mean())
plt.show()
```

# 3. Sex

```python
titanic_df.groupby(['Survived', 'Sex'])['Survived'].count()
```

```python
#sns.catplot

# col으로 case 분류
# count를 기준으로 그래프 표현

sns.catplot(x='Sex', col = 'Survived', kind = 'count', data = titanic_df)
plt.show()
```

# 4. Age
- Remind ; 결측치 존재

```python
titanic_df.describe()['Age']
```

```python
# Survived 1, 0과 Age의 경향성
# figure -> axis -> plot

fig, ax = plt.subplots(1, 1, figsize = (10, 5))
sns.kdeplot(x = titanic_df[titanic_df.Survived == 1]['Age'], ax = ax)
sns.kdeplot(x = titanic_df[titanic_df.Survived == 0]['Age'], ax = ax)

plt.legend(['Survived', 'Dead'])

plt.show()
```

# Appendix 1. Sex + Pclass vs Survived

```python
sns.catplot(x = 'Pclass', y = 'Survived', hue = 'Sex', kind = 'point', data = titanic_df)
plt.show()
```
